```python
# Importing the ConfigProto class from TensorFlow's v1 compatibility module
from tensorflow.compat.v1 import ConfigProto

# Importing the InteractiveSession class from TensorFlow's v1 compatibility module
from tensorflow.compat.v1 import InteractiveSession

# Creating a ConfigProto object, which is used to configure TensorFlow sessions
config = ConfigProto()

# Setting the fraction of the GPU memory to be allocated to the TensorFlow process to 50%
config.gpu_options.per_process_gpu_memory_fraction = 0.5

# Allowing dynamic GPU memory allocation, meaning TensorFlow will allocate memory on the GPU as needed
config.gpu_options.allow_growth = True

# Creating an interactive TensorFlow session with the specified configuration
# This session will be used for running TensorFlow operations
session = InteractiveSession(config=config)
```

```python
# import the libraries as shown below

from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet50 import ResNet50
#from keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
# re-size all the images to this
IMAGE_SIZE = [224, 224]

train_path = 'Data/train'
valid_path = 'Data/test'
```

```python
# Import the Vgg 16 library as shown below and add preprocessing layer to the front of VGG
# Here we will be using imagenet weights

resnet = ResNet50(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
# don't train existing weights
for layer in resnet.layers:
    layer.trainable = False
```

```python
  # useful for getting number of output classes
folders = glob('Data/train/*')
```

```python
# our layers - you can add more if you want
x = Flatten()(resnet.output)
```

```python
prediction = Dense(len(folders), activation='softmax')(x)

# create a model object
model = Model(inputs=resnet.input, outputs=prediction)
```

```python
# view the structure of the model
model.summary()
```

```python
# tell the model what cost and optimization method to use
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
# Use the Image Data Generator to import the images from the dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# Make sure you provide the same target size as initialied for the image size
training_set = train_datagen.flow_from_directory('Data/train',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory('Data/test',
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
# fit the model
# Run the cell. It will take some time to execute
r = model.fit_generator(
  training_set,
  validation_data=test_set,
  epochs=20,
  steps_per_epoch=len(training_set),
  validation_steps=len(test_set)
)
```
